Handle Missing Value

```python
#Show count of NA in each features
print(df.isna().sum())
```

```python
for column in df.columns:
    if df[column].dtype == 'object':
        count = (df[column] == '-').sum()
        if count > 0:
            print(f"Column '{column}': {count} occurrences of '-'")

import numpy as np

print("Total NaN before replacing:")
print(df.isna().sum())
print()

# Replace '-' with NaN
df.replace(['-'], np.nan, inplace=True)

print("Total NaN after replacing '-':")
print(df.isna().sum())
```

Remove Outliers

```python
labels_to_remove = ['FileDownload']
df = df[~df['label'].isin(labels_to_remove)]
df['label'].value_counts()
```

One Hot Encoding

```python
#Perform One-Hot Encoding on proto and conn_state set as binary

df = pd.get_dummies(df, columns=['proto'],dtype=int)
df = pd.get_dummies(df, columns=['conn_state'],dtype=int)
```

```python
df.head()
```

Removed Skewedness

```python
df.shape
```

```python
numeric_cols = ['duration', 'orig_bytes', 'resp_bytes']

df[numeric_cols] = df[numeric_cols].replace('-', 0.0)

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0.0)

print(df[numeric_cols].dtypes)
```

```python
df = df[
    ~(
        (df['duration'] == 0) &
        (df['orig_bytes'] == 0) &
        (df['resp_bytes'] == 0) &
        (df['missed_bytes'] == 0) &
        (df['orig_pkts'] == 1) &
        (df['resp_pkts'] == 0) &
        (df['proto_tcp'] == 1) &
        (df['conn_state_S0'] == 1)
    )
]

mask = (
    (df['duration'] == 0) &
    (df['orig_bytes'] == 0) &
    (df['resp_bytes'] == 0) &
    (df['missed_bytes'] == 0) &
    (df['orig_pkts'] == 1) &
    (df['resp_pkts'] == 0) &
    (df['proto_tcp'] == 1) &
    (df['conn_state_S0'] == 1)
)

print("Rows removed:", mask.sum())
df = df[~mask]
```

Feature Engineering

```python
df.shape
```

```python
print(df.label.value_counts())
```

```python
#Connection Rate (Flows Per Second)
df['ts'] = pd.to_datetime(df['ts'], unit='s')
df = df.sort_values('ts')

# Set the index to timestamp for rolling window calculations
df = df.set_index('ts')

# Group by Source IP to analyze behavior per device
gb = df.groupby('id.orig_h')

# Feature A: Connection Rate (Flows per second)
df['conn_rate_1s'] = gb['uid'].rolling('1s').count().reset_index(0, drop=True)

# Feature B: Port Variety
df['unique_ports_1m'] = gb['id.resp_p'].rolling('60s').apply(lambda x: len(set(x))).reset_index(0, drop=True)

# Feature C: Inter-Arrival Time (IAT)
df['iat'] = gb.apply(lambda x: x.index.to_series().diff().dt.total_seconds()).reset_index(0, drop=True).fillna(0)

# Feature D: Average Packet Size (using ip_bytes for more detail)
df['avg_orig_pkt_size'] = df['orig_ip_bytes'] / (df['orig_pkts'] + 1e-5)
```

Undersampling the Major Class

```python
from sklearn.utils import resample

# Separate majority and minority classes
df_majority = df[df.label == 'PartOfAHorizontalPortScan']
df_minority = df[df.label != 'PartOfAHorizontalPortScan']

# Undersample majority class
df_majority_undersampled = resample(df_majority,
                                 replace=False,    # sample without replacement
                                 n_samples=300000, # to match desired number
                                 random_state=42) # reproducible results

# Combine minority class with undersampled majority class
df = pd.concat([df_majority_undersampled, df_minority])

# Display new class counts
print(df.label.value_counts())
```
